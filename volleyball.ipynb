```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("VNL2023.csv")
```

```python
numeric_col = df.select_dtypes(include=["int","float"]).columns
corr_mat = df[numeric_col].corr()
print(corr_mat)
```

```python
sns.heatmap(corr_mat, annot=True, cmap="coolwarm", linewidths=5)
plt.title("Correlation matrix heat map")
plt.show()
```

```python
position_counts = df["Position"].value_counts()
position_counts
```

```python
plt.pie(position_counts, labels=position_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Distribution of Positions")
plt.show()
```

```python
avg_attack_country = df.groupby("Country")["Attack"].mean()
avg_attack_country.sort_values(ascending=False)
```

```python
avg_attack_country.sort_values(ascending=False).plot(kind="bar")
plt.title("Average Attack by Country")
plt.xlabel("Country")
plt.ylabel("Average Attack")
plt.show()
```

```python
avg_serve_by_age = df.groupby("Age")["Serve"].mean()
avg_serve_by_age.sort_values(ascending=False)
```

```python
df.groupby(["Country", "Position"])["Attack"].mean().reset_index().sort_values(ascending=False,by="Attack" )
```

```python
df.groupby("Country")["Dig"].sum()
```

```python
plt.scatter(df["Block"], df["Receive"])
plt.title("Scatter Plot : Block vs Receive")
plt.xlabel("Block")
plt.ylabel("Recieve")
plt.show()
```

```python
sns.boxplot(x=df["Serve"])
plt.title("Box Plot: Dostribution of Serve")
plt.xlabel('Serve')
plt.show()
```

```python
plt.hist(df["Age"], bins =20, color='skyblue', edgecolor='black')
plt.title("Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
avg_attack_by_position = df.groupby("Position")["Attack"].mean()
avg_attack_by_position.plot(kind='bar', color="red")
plt.title("Average Attack by Position")
plt.xlabel("Position")
plt.ylabel("Average Attack")
plt.show()
```

```python
serve_trend_by_age = df.groupby("Age")["Serve"].mean()
serve_trend_by_age.plot(kind="line", marker="o", linestyle="-", color="orange")
plt.title("Serve Trend Over Age Group")
plt.xlabel("Age")
plt.ylabel("Average Serve")
plt.show()
```

```python
total_attack_block_by_country = df.groupby("Country")[["Attack", "Block"]].sum()
total_attack_block_by_country.plot(kind="bar", stacked=True, colormap="viridis")
plt.title("Total Attack and Block by Country")
plt.xlabel("country")
plt.ylabel("Total Value")
plt.show()
```
